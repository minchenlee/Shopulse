```python
import os
import pprint
import asyncio
import aiohttp
from dotenv import load_dotenv
import jieba
from openai import OpenAI
```

### 文本處理

```python
folder_name = "./result/如何挑選電視？"
```

#### 載入文本

```python
# Sample text in Chinese
with open(f"{folder_name}/sum_result.txt", 'r') as f:
    text = f.read()
    
# Split text into sentences (a more sophisticated method might be needed for complex texts)
sentences = text.split('。')
```

#### 進行 embedding

```python

load_dotenv()  # import the environment variables
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

def get_embeddings(sentence):
    response = client.embeddings.create(
    model="text-embedding-ada-002",
    input=sentence,
    encoding_format="float"
    )
    print(response.data[0].embedding)
    return response.data[0].embedding

# For loop to get embeddings for each sentence
sentences_embeddings = []
for sentence in sentences:
    sentences_embeddings.append(get_embeddings(sentence))
```

```python
print(sentences_embeddings[0])
```

#### 存 embedding

```python
with open(f'{folder_name}/embeddings/embeddings.txt', 'w') as file:
    file.write(str(sentences_embeddings))
```

#### 載入 embeddings

```python
# import embedding.txt
with open(f"{folder_name}/embeddings/embeddings.txt", 'r') as f:
    sentences_embeddings = f.read()

# convert string to list, it is 2D list
sentences_embeddings = eval(sentences_embeddings)
```

```python
print(sentences_embeddings[0])
```

```python
import numpy as np
from sklearn.cluster import KMeans

# Assuming you have multiple sentences and their embeddings
number_of_clusters = 8  # Adjust as needed

# Use KMeans clustering
kmeans = KMeans(n_clusters=number_of_clusters)
kmeans.fit(sentences_embeddings)

# The cluster assignment for each sentence
clusters = kmeans.labels_
```

```python
grouped_sentences = [[] for _ in range(number_of_clusters)]
for sentence, cluster_id in zip(sentences, clusters):
    grouped_sentences[cluster_id].append(sentence)
```

```python
for i, cluster in enumerate(grouped_sentences):
    print(f"Cluster {i}:")
    for sentence in cluster:
        print(sentence)
    print("--------------------")
```
